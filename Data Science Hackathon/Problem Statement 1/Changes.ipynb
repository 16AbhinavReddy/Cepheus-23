```python
import pandas as pd
```

```python
data = pd.read_csv("microbes.csv")
pd.set_option('display.max_columns',500)
```

```python
data
```

```python
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
import seaborn as sbn
```

```python
sbn.countplot(x="microorganisms",data=data)
```

```python
data["microorganisms"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
label = data["microorganisms"].to_numpy().reshape(-1,1)
data["microorganisms"] = encoder.fit_transform(label)
data
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x = data.drop(["microorganisms"],axis=1)
y = data["microorganisms"]
```

```python
sm = SMOTE(random_state=71)
x_new, y_new = sm.fit_resample(x,y)
```

```python
new_data = x_new
new_data["microorganisms"] = y_new
```

```python
new_data
```

```python
new_data1 = new_data.sample(frac=1)
new_data1
```

```python
new_data1 = new_data1.reset_index()
new_data1 = new_data1.drop(["index"],axis=1)
```

```python
new_data1
```

```python
new_data1["microorganisms"] = new_data1["microorganisms"].map({0: "Cruor", 1: "Myaree", 2: "Elixoria", 3: "Onyxium", 4: "Raizovox", 5: "Flitter", 6: "Velvette", 7: "Gelledius", 8: "Potentiopus", 9: "Charmayanne"})
```

```python
new_data1
```

```python
new_data1["microorganisms"].unique()
```

```python
new_data1 = new_data1.rename(columns={'Solidity': 'Compactness', 'Orientation':'Direction', 'Extrema' : 'Points of Extremum', 'microorganisms': 'Bacteria', 'BoundingBox1': 'Bordered Area 1', 'BoundingBox2': 'Bordered Area 2', 'BoundingBox3': 'Bordered Area 3', 'BoundingBox4': 'Bordered Area 4', 'FilledArea': 'Filled Area', 'ConvexHull1':'Convex Hull 1', 'ConvexHull2':'Convex Hull 2', 'ConvexHull3': 'Convex Hull 3', 'ConvexHull4': 'Convex Hull 4', 'MajorAxisLength': 'Length of Major Axis', 'MinorAxisLength': 'Length of Minor Axis', 'ConvexArea': 'Convex Area', 'Centroid1': 'Centroid 1', 'Centroid2': "Centroid 2" })
new_data1
```

```python
names = []
for i in range(len(new_data1)):
  name = "CEPHEUS_CIPHERS_0" + str(10101010 + i)
  names.append(name)
```

```python
Names = pd.DataFrame(names)
```

```python
new_data1["Testing Sample ID"] = Names
new_data1
```

```python
new_data1["Equivalent Radius"] = new_data1["EquivDiameter"]/2
new_data1["Diameter"] = new_data1["raddi"] * 2
new_data1 = new_data1.drop(["EquivDiameter", "raddi", "EulerNumber"], axis=1)
new_data1
```

As Compactness/ soldity is the ratio of area of the object and the convex area feature, What I will do is remove area feature from the dataset. As we will provide what compactness is about, participants should create area feature with the help of this definition in the feature engineering part. Similarly I will remove eccentricity feature.

```python
new_data1 = new_data1.drop(["Area", "Eccentricity"], axis=1)
new_data1
```

```python
new_data1 = new_data1.sample(frac=1, axis=1)
new_data1
```

```python
names_data = new_data1["Testing Sample ID"]
length_data = new_data1[["Length of Major Axis", "Length of Minor Axis"]]
Bounding_data = new_data1[["Bordered Area 1", "Bordered Area 2", "Bordered Area 3", "Bordered Area 4"]]
Centroid_data = new_data1[["Centroid 1", "Centroid 2"]]
Convex_Hull_data = new_data1[["Convex Hull 1", "Convex Hull 2", "Convex Hull 3", "Convex Hull 4"]]
```

```python
rem_data = new_data1[["Perimeter", "Filled Area", "Convex Area","Equivalent Radius", "Extent", "Points of Extremum", "Compactness", "Diameter", "Direction"]]
```

```python
label = new_data1["Bacteria"]
```

```python
Final_Data = pd.concat([names_data, rem_data, length_data, Bounding_data, Centroid_data, Convex_Hull_data,], axis=1)
Final_Data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(Final_Data, label, test_size=0.25, random_state=101)
```

```python
x_train
```

```python
Train = pd.concat([x_train, y_train],axis=1)
Test = pd.concat([x_test, y_test], axis=1)
```

```python
Train = Train.reset_index()
Test = Test.reset_index()
```

```python
Train = Train.drop("index", axis=1)
```

```python
Test_Answers = Test["Bacteria"]
Test = Test.drop(["Bacteria","index"], axis=1)
Test
```
