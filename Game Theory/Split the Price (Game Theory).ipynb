```python
ID = ['CEPHEUS369', 'CEPHEUS828', 'CEPHEUS640', 'CEPHEUS438', 'CEPHEUS597', 'CEPHEUS117', 'CEPHEUS583', 'CEPHEUS841', 'CEPHEUS241', 'CEPHEUS697', 'CEPHEUS977', 'CEPHEUS273', 'CEPHEUS207', 'CEPHEUS795', 'CEPHEUS262', 'CEPHEUS938', 'CEPHEUS824', 'CEPHEUS529', 'CEPHEUS456', 'CEPHEUS940', 'CEPHEUS939', 'CEPHEUS644', 'CEPHEUS427', 'CEPHEUS136', 'CEPHEUS152', 'CEPHEUS899', 'CEPHEUS101', 'CEPHEUS905', 'CEPHEUS781', 'CEPHEUS164']
new_data = data.copy()
l = data["Enter your Wallet ID provided"].tolist()
for i in range(len(data)):
    if l[i] not in ID:
        new_data = data.drop(i, axis=0)
data = new_data
```

```python
data_part = data_part.sort_values(by= "Enter the Money Holder Number you received", ascending=1)
data_part = data_part.reset_index()
data_part = data_part.drop(["index"],axis=1)
data_part
```

```python
std_data = data_part["Enter the Money Holder Number you received"].to_list()
std_data_new = data["Enter your Wallet ID provided"].to_list()
for i in range(len(data_part)):
    if std_data[i] not in std_data_new:
        data.loc[len(data.index)] = [0 ,std_data[i], 0]
```

```python
data
```

```python
data = data.sort_values(by= "Enter your Wallet ID provided", ascending=1)
data = data.reset_index()
data = data.drop(["index"],axis=1)
data
```

```python
Inputs = data["Choose a Price"].to_list()
print(Inputs)
```

```python
values = set()
for i in Inputs:
    values.add(i)
options = {}
for i in Inputs:
    if i not in options:
        options[i] = 1
    else:
        options[i] += 1
output = {}
for i in values:
    output[i] = i/options[i]
print(output)
```

```python
final_output = []
for i in Inputs:
    final_output.append(output[i])
print(final_output)
```

```python
FinalOutput = pd.DataFrame(final_output)
FinalOutput
```

```python
data["output"] = FinalOutput[0]
data
```

```python
data = data.sort_values(by="output", ascending=0)
data.to_csv("Final.csv")
```

```python
Updates_Scores = data_part
Updates_Scores["Amount"] = Updates_Scores["Amount"] + final_output
Updates_Scores.to_csv("ParticipantsData.csv")
```
